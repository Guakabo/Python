# BMI Calculator
This program will show distribution of the Body Mass Index (BMI) based on data. The mean, standard deviation, minimum, maximum, and median will be calculated by this program. The program will also look at the BMI data and determine the amount of underweight and obese people in both male and female population.

Input: two data files

Process:
1. Open files
2. Turn files to four separate arrays
3. Convert pounds to kilograms
4. Converts centimeters to meters
5. Calculate BMI
6. Combine male and female height and weight into a single array
7. Output all distributions
8. Output number of underweight and obese males and females

Output: male height, weight and BMI data, female height, weight and BMI data. Combined height and weight data as well as number of obese and underweight males and females

Date:11/10/2020

```python
#Importing numpy and pandas modules
import numpy as np

import pandas as pd

#Functions to calculate kilograms, meters, and bmi
def pounds_to_kilogram_converter(pounds):
    kilogram = pounds /2.2046
    return kilogram

def centimeter_to_meter_converter(centimeter):
    meter = centimeter/100
    return meter
    
def bmi_calculator(kilograms,meters):
    bmi = kilograms/(meters**2)
    return bmi

#Pandas reads the csv files
maleRaw = pd.read_csv('measurements-male.csv')

femaleRaw = pd.read_csv('measurements-female.csv')

#Create weight and height arrays for males and females
maleWeights = np.array(maleRaw['Weight (lb)'])

femaleWeights = np.array(femaleRaw['Weight (lb)'])

maleHeights = np.array(maleRaw['Height (cm)'])

femaleHeights = np.array(femaleRaw['Height (cm)'])

#Convert the weight arrays using function
maleWeightKg= pounds_to_kilogram_converter(maleWeights)

femaleWeightKg= pounds_to_kilogram_converter(femaleWeights)

#Convert the height arrays using function
maleHeightMeters= centimeter_to_meter_converter(maleHeights)

femaleHeightMeters= centimeter_to_meter_converter(femaleHeights)

#Calculate BMI
maleBMI= bmi_calculator(maleWeightKg,maleHeightMeters)

femaleBMI= bmi_calculator(femaleWeightKg,femaleHeightMeters)

#Male Heights distribution
print('Here is the distribution for male heights in meters: ')
print("Male Mean height:       ", maleHeightMeters.mean() )
print("Male Standard deviation:", maleHeightMeters.std() )
print("Male Minimum height:    ", maleHeightMeters.min() )
print("Male Maximum height:    ", maleHeightMeters.max() )
print("Male Median height:     ", np.median(maleHeightMeters) )

#Female Heights distribution
print('\nHere is the distribution for female heights in meters: ')
print("Female Mean height:       ", femaleHeightMeters.mean() )
print("Female Standard deviation:", femaleHeightMeters.std() )
print("Female Minimum height:    ", femaleHeightMeters.min() )
print("Female Maximum height:    ", femaleHeightMeters.max() )
print("Female Median height:     ", np.median(femaleHeightMeters) )

#Combined Heights distribution
combinedHeights = np.concatenate((femaleHeightMeters,maleHeightMeters))
print('\nHere is the overall height distribution in meters: ')
print("Mean height:       ", combinedHeights.mean() )
print("Standard deviation:", combinedHeights.std() )
print("Minimum height:    ", combinedHeights.min() )
print("Maximum height:    ", combinedHeights.max() )
print("Median height:     ", np.median(combinedHeights) )

#Male Weights distribution
print('\nHere is the distribution for male weights in kilograms: ')
print("Male Mean weight:       ", maleWeightKg.mean() )
print("Male Standard deviation:", maleWeightKg.std() )
print("Male Minimum weight:    ", maleWeightKg.min() )
print("Male Maximum weight:    ", maleWeightKg.max() )
print("Male Median weight:     ", np.median(maleWeightKg) )

#Female Weights distribution
print('\nHere is the distribution for female weights in kilograms: ')
print("Female Mean weight:       ", femaleWeightKg.mean() )
print("Female Standard deviation:", femaleWeightKg.std() )
print("Female Minimum weight:    ", femaleWeightKg.min() )
print("Female Maximum weight:    ", femaleWeightKg.max() )
print("Female Median weight:     ", np.median(femaleWeightKg) )

#Combined Weights distribution
combinedWeights = np.concatenate((femaleWeightKg,maleWeightKg))
print('\nHere is the overall height distribution in kilograms: ')
print("Mean height:       ", combinedWeights.mean() )
print("Standard deviation:", combinedWeights.std() )
print("Minimum height:    ", combinedWeights.min() )
print("Maximum height:    ", combinedWeights.max() )
print("Median height:     ", np.median(combinedWeights) )

#Male BMI distribution
print('\nHere is the distribution for male BMIs: ')
print("Male Mean BMI:       ", maleBMI.mean() )
print("Male Standard deviation:", maleBMI.std() )
print("Male Minimum BMI:    ", maleBMI.min() )
print("Male Maximum BMI:    ", maleBMI.max() )
print("Male Median BMI:     ", np.median(maleBMI) )

#Female BMI distribution
print('\nHere is the distribution for female BMIs: ')
print("Female Mean BMI:       ", femaleBMI.mean() )
print("Female Standard deviation:", femaleBMI.std() )
print("Female Minimum BMI:    ", femaleBMI.min() )
print("Female Maximum BMI:    ", femaleBMI.max() )
print("Female Median BMI:     ", np.median(femaleBMI) )

#Number of underweight people separated by gender
print('\n Here is the amount of underweight males: ',np.sum(maleBMI<18.5))

print('\n Here is the amount of underweight females: ',np.sum(femaleBMI<18.5))

#Number of obese people separated by gender
print('\n Here is the amount of obese males: ',np.sum(maleBMI>=30))

print('\n Here is the amount of obese females: ',np.sum(femaleBMI>=30))
```
