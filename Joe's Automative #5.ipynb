# Joe's Automotive
The purpose of this program is to create a program that can inform the customer of the services offered in Joe's automotive 
as well as keeping a list of products selected by the customers. This list of services will be checked with our list of
makes to see if the customer's car is eligible for a discount.

Input: Desired service from the provided options, make of your vehicle

Output: Display error message if the service is out of range, display if service has been selected already (duplicate),
and will display the total desired services by the customer along with their discount (if eligible)

Author: Christian Sierra

Date: 10/06/2020

```python
#Create services Tuple
availableServices = ('Oil Change','Lube Job','Radiator Flush','Transmission Flush','Inspection','Muffler Replacement',
'Tire Rotation')

#Create Makes Tuple
discountedMakes = ('Acura','Honda','Infiniti','Lexus','Mazda','Mitsubishi','Nissan','Subaru','Suzuki','Toyota')

#Create empty list for the services selected by the user
desiredServices = []

#Displays all services
print('Welcome! Joe\'s automotives provides the following services: \n 1. Oil Change \n 2. Lube Job \n '
'3. Radiator Flush \n 4. Transmission Flush \n 5. Inspection \n 6. Muffler Replacement \n 7. Tire Rotation')

#Takes floats and converts them to integer
service = float(input('Please select the number for the service items you need. Enter -1 to stop: '))
service = int(service)
#Set bolean value for the while loop
switchOn = True

#While loop will run while both conditions are true
while service != -1 and switchOn == True:
    #This nested while loop validates numbers from 1 to 7 as requested by the program
    while service > 7 or service <= 0:
        service = float(input('It appears the number entered is out of range! Enter a number between 1 and 7 to select'
        ' from the options provided above! Enter -1 to stop: '))
        service = int(service)
        #Since if -1 is entered the loop will no stop, this condition ensures -1 ends the loop.
        if service == -1:
            #Change loop condition and breaks out of the loop
            switchOn = False
            break
        else:
            pass
    #Ensures the rest of the loops does not run if -1 was entered in the validation step
    if switchOn == False:
        break
    #If the input was valid, this code will execute and lower the input to the correct index value
    service = int(service)
    service -= 1
    #Verifies if the services has been selected already
    if availableServices[service] in desiredServices:
        print('The service "'+availableServices[service]+'" is already in the cart.')
    #Adds the service to the list
    else:
        print('The service "'+availableServices[service]+ '" has been added to the cart')
        desiredServices += [availableServices[service]]
    #Input for next service or to terminate the program by entering -1
    service = float(input('Please select the number for the service items you need. Enter -1 to stop: '))
    service = int(service)

#In case -1 is entered from the beginning, this code will trigger the desired message
if service == -1 and len(desiredServices) == 0:
        print('You did not select any service. Sorry we cannot make the reservation for you.')
#If the input is not -1, it will ask for the make
else:
    make = input('Please enter the make of your car: ')
    #The make is capitalized so it matches the values on the tuple
    make = make.capitalize()
    print('Thank you for making the reservation! The services you selected are as following: ')
    for i in desiredServices:
        print(i)
    #Queries the makes tuple to see if the make is eligible for the discount
    if make in discountedMakes:
        print('We will provide a 10% discount for your '+make+' vehicle!')
    #Triggers if the make is not on the discount list
    else:
        print('These services will be provided for your '+make+' !')
```
