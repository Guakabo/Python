# Room Rate Checker
This application will check room number pricing based on the selected room preferences. The program will ensure the proper format and the proper pricing depending on the amenities located in that room. 

Input: 4 digit room ID, user choice for another room

Process:

1. Define function
2. Create empty dictionary
3. While loop iteration to ensure proper input
4. Value gets stripped of extra spaces and formatted to lower case characters
4. Proper input gets stored as key/value pair into the dictionary
5. Prompt user to add more rooms or stop
6. Display the store key/value pairs 

Output: Error messages in case of improper input, room ID and cost

```python
#Function to calculate room rate
def rate_calculator (roomId):
    #This is set as a variable in case it needs to be changed in the future.
    roomBaseValue = 60
    #Below are all the possible valid values for hotel room IDs. They all return an integer value back when called
    if roomId[0].lower() == 'l' and roomId[-1].lower() == 'k' and int(roomId[1])>3:
        return roomBaseValue + 60
    elif roomId[0].lower() == 'l' and roomId[-1].lower() == 's' and int(roomId[1])>3:
        return roomBaseValue + 50
    elif roomId[0].lower() == 'n' and roomId[-1].lower() == 's' and int(roomId[1])>3:
        return roomBaseValue + 30
    elif roomId[0].lower() == 'n' and roomId[-1].lower() == 'k' and int(roomId[1])>3:
        return roomBaseValue + 40
    elif roomId[0].lower() == 'l' and roomId[-1].lower() == 's':
        return roomBaseValue + 20
    elif roomId[0].lower() == 'l' and roomId[-1].lower() == 'k':
        return roomBaseValue + 30
    elif roomId[0].lower() == 'n' and roomId[-1].lower() == 'k':
        return roomBaseValue + 10
    elif roomId[0].lower() == 'n' and roomId[-1].lower() == 's':
        return roomBaseValue

#Empty dictionary which the user will fill with their input
roomsAndPrices = {}

#Used a boolean to set the while loop condition
isValid = True
while isValid:
    roomNum = input('Please enter the room number: ')
    roomNum = roomNum.strip().lower()
    
    #This nested while loop serves to validate the input from the user based on program specifications
    while not len(roomNum) == 4 or (not roomNum.startswith('l') and not roomNum.startswith('n')) or (not roomNum.endswith('k') and not roomNum.endswith('s')) or not int(roomNum[1:3].isnumeric()): 
        print('The room number should be 4 alphanumeric digits. \n The first character should be either n or l.'
        '\n The last character should be either k or s. \n The second and third character should be numbers.')
        roomNum = input('Please enter again! ')
        roomNum = roomNum.strip().lower()
    
    #Adds the room number ID as a key with the return value of the function 'rate_calculator' as a value pair
    roomsAndPrices[roomNum] = rate_calculator(roomNum)
    
    #Displays the value of the room inquired by the user
    print('The room '+roomNum+' is $'+ str(rate_calculator(roomNum))+ ' per night! ')
    
    #Asks the user if they want to add more rooms to the dictionary
    anotherRoom = input ('Would you like to add another room? Y/N ')
    
    #Validation input for the while loop in order to ensure a 'y' or 'n' answer.
    while not anotherRoom.lower() == 'y' and not anotherRoom.lower() == 'n':
        anotherRoom = input ('Please enter Y/N ')
    
    #Conditional statement, if user wishes to add more rooms, 'isValid' will remain True allow the loop to restart
    if anotherRoom.lower() == 'y':
        isValid = True
    #If the user does not want to add another room, 'isValid' is changed to False, causing the loop to finalize
    else:
        isValid = False

#Displays the dictionary to the user
print('Here are all the rooms and the rate for each room you just checked: ')

#This for loop iterates through all the items inside the dictionary and displays them as a key/value pair
for k, v in roomsAndPrices.items():
    
    #This formats the display
    print(str(k).ljust(15, '.') + '$'+ str(v).rjust(1,'.'))
```
